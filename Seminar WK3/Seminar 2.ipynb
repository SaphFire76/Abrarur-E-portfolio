```python
import socket

def get_ip_address(website_url):
    try:
        ip_address = socket.gethostbyname(website_url)
        print(f"The IP address of {website_url} is {ip_address}")
    except socket.gaierror:
        print(f"Unable to get the IP address for {website_url}")

# Example usage
#website = input("Enter the website URL (without 'https://'): ")
get_ip_address('www.google.com')
get_ip_address('www.amazon.com')
get_ip_address('www.bbc.co.uk')
```

```python
#Exercise 1#

import subprocess

def tracert(domain):
    try:
        result = subprocess.run(["tracert", domain], capture_output=True, text=True)
        print(result.stdout)
    except FileNotFoundError:
        print("tracert command not found. Make sure it's available.")
    except Exception as e: # Catching general exceptions for now
        print(f"An error occurred: {e}")

#domain = input("Enter the website or IP address: ")
tracert('www.google.com')
tracert('www.amazon.com')
tracert('www.bbc.co.uk')
```

#### #Exercise 3 and 4#

The 3 domains used were:
- www.google.com
- www.amazon.com
- www.bbc.co.uk

Out of these 3 domains amazon.com was the least transparent with the identity of the routers taken en route to its destination. Many of the routers for amazon.com, and some for google.com and bbc.co.uk, destroyed the packets and did not send back the ICMP 'Time Exceeded' message back, therefore tracert was unable to produce the IP address and RTT of the router. After discussing the results with the T.A. in the seminar I understood that routers may destroy packets and not return a message as a form of security. Hackers with malicious intent may take advantage of the information collected by tracert to indetify vulnerabilities in the route and exploit these.

Bottlenecks are points where the flow of data is impaired or stopped entirely. The bottlenecks discovered en route to the 3 domains are as follows:
- www.google.com: The router '100.127.248.51' was a bottleneck as the RTT on its second packet was 10ms, which is the highest RTT recorded en route to google.com.
- www.amazon.com: It's hard to assess which routers acted as bottlenecks en route to amazon.com, as most of the routers were not identified and did not send back an RTT. However, from the limited information available, the router '100.71.208.1' had the highest RTT and is therefore the bottleneck.
- www.bbc.co.uk: En route to bbc.co.uk the biggest bottleneck was the router '195.66.224.88' with an average RTT of 14ms. The router '100.71.208.1' follows closely behind with an RTT of 9.7ms.

```python
import socket

# Create a socket object
client_socket = socket.socket(socket.AF_INET, socket.SOCK_STREAM)

# Define the server address and port
server_address = ('www.bbc.co.uk', 80)

# Connect to the server
client_socket.connect(server_address)

# Send an HTTP GET request
request = "GET / HTTP/1.1\r\nHost: www.bbc.co.uk\r\n\r\n"
client_socket.send(request.encode())

# Receive the response
response = client_socket.recv(4096)
print(response.decode())

# Close the socket
client_socket.close()
```

```python
import requests

url = 'https://jsonplaceholder.typicode.com/posts'
data = {
    "title": "Sample Post",
    "body": "This is an example post body.",
    "userId": 1
}

response = requests.post(url, json=data)
print(f"Status Code: {response.status_code}")
print("Response Body:", response.json())
```

```python
url = 'https://jsonplaceholder.typicode.com/posts/1'
updated_data = {
    "id": 1,
    "title": "Updated Title",
    "body": "This post content has been updated.",
    "userId": 1
}

response = requests.put(url, json=updated_data)
print(f"Status Code: {response.status_code}")
print("Updated Resource:", response.json())
```
