```python
# Below is the code for an application that sends and receives messages between clients on the same network.
# I created this in the seminar and used it to talk to some classmates.
# The application worked for the most part, but it was not perfect.
# The biggest problem was that, to receive a message you had to press enter.
# So, if someone had sent a message on the network, you would only know after pressing the enter.
# Except that and a few refinements, the application worked as intended, and we were able to communicate while on the same network.
# Please ignore the contents of the messages exchanged.

import socket
import threading

# Create UDP socket
server_socket = socket.socket(socket.AF_INET, socket.SOCK_DGRAM)
server_socket.bind(('172.20.10.4', 65433))

print("UDP Server is ready to send and receive data...")

# Function to handle receiving messages
def receive_messages():
    while True:
        data, client_address = server_socket.recvfrom(2048)
        print(f"\nReceived from {client_address}: {data.decode()}")

# Function to handle sending messages
def send_messages():
    while True:
        message = input("\nEnter message to send: ")
        message = 'Abrarur: '+ message
        server_socket.sendto(message.encode(), ('172.20.10.8', 65433))
        print(message)

# Create threads for sending and receiving
receive_thread = threading.Thread(target=receive_messages)
send_thread = threading.Thread(target=send_messages)

# Start the threads
receive_thread.start()
send_thread.start()

# Join threads to main thread
receive_thread.join()
send_thread.join()
```
